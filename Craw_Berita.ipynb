## Import Library

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## Crawling Data dengan BS4 dan Request

```python
#Inisialisasi Link URL
url = 'https://news.detik.com/indeks'

headers={
    'user-agent': 'Mozilla/5.0 (iPhone; CPU iPhone OS 13_2_3 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/13.0.3 Mobile/15E148 Safari/604.1'
    }

#inisialisasi list untuk memasukkan hasil crawling kedalam list
listJudul = []
listIsi = []

#proses perulangan untuk crawling data

#perulangan halaman web dari halaman 1 sampe 4
for page in range(1,4):
  req = requests.get(url+'/'+str(page), headers=headers)
  soup = BeautifulSoup(req.text, 'html.parser')
  items = soup.findAll('h3','media__title')
  for it in items:
    try: link = it.find('a')['href']
    except : link=''
    try: title = it.find('div').text.replace('\n','')
    except : title=''

    if title != '':
      listJudul.append(title)

    if link != '':
      req2 = requests.get(str(link), headers=headers)
      soup2 = BeautifulSoup(req2.text, 'html.parser')
      items2 = soup2.findAll("article", "detail")
      for it2 in items2:
        try : berita = it2.find('p').text
        except : berita=''

        if berita != '':
          listIsi.append(berita)

df_Judul = pd.DataFrame(listJudul, columns=["Judul Berita"])
df_Isi = pd.DataFrame(listIsi, columns=["Isi Berita"])

data = pd.concat([df_Judul, df_Isi], axis=1)
data = data.loc[0:50]
data
```
