## Import Library Twint 

```python
import twint
import pandas
import nest_asyncio
nest_asyncio.apply()
```

`## Crawling Data Twitter - Ganjar Pranowo`

```python
c = twint.Config()
c.Search = "ridwankamil"
c.Since = "2023-01-01"
c.Until = "2023-02-16"
c.Limit = 500
c.Pandas = True
c.Lang = "id"

twint.run.Search(c)
```

## Membuat dataframe untuk hasil crawling data


```python
data = twint.storage.panda.Tweets_df
df = pandas.DataFrame(data, columns=['date', 'username', 'tweet'])
df
```
